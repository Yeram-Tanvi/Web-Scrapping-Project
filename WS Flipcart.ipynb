```python
import numpy as np 
import pandas as pd 
import warnings 
warnings.filterwarnings('ignore')
import requests 
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Referer': 'https://www.google.com/',
    'DNT': '1',
    'Upgrade-Insecure-Requests': '1'
}

name = []
rating = []
review = []
price = []
og_price = []
discount = []
des = []

for i in range(2,45):
    url = "https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_2_na_na_na&as-pos=2&as-type=HISTORY&suggestionId=laptop%7CLaptops&requestId=247fe4a2-d5ff-425b-bdb6-0c65951d788a&page="+str(i)

    webpage = requests.get(url, headers=headers)
    soup = BeautifulSoup(webpage.text, "html.parser")

    boxes = soup.find_all('div', class_="QSCKDh dLgFEE")

    for box in boxes:

        cards = box.find_all('a', class_="k7wcnx")

        for card in cards:

            name_tag = card.find('div', class_="RG5Slk")
            rating_tag = card.find('div', class_="MKiFS6")
            review_tag = card.find('span', class_="PvbNMB")
            price_tag = card.find('div', class_="hZ3P6w DeU9vF")
            og_price_tag = card.find('div', class_="kRYCnD gxR4EY")
            discount_tag = card.find('div', class_="HQe8jr")
            des_tag = card.find('ul', class_="HwRTzP")

            name.append(name_tag.text.strip() if name_tag else "N/A")
            rating.append(rating_tag.text.strip() if rating_tag else "N/A")
            review.append(review_tag.text.strip() if review_tag else "N/A")
            price.append(price_tag.text.strip() if price_tag else "N/A")
            og_price.append(og_price_tag.text.strip() if og_price_tag else "N/A")
            discount.append(discount_tag.text.strip() if discount_tag else "N/A")
            des.append(des_tag.text.strip() if des_tag else "N/A")

    print("Products:", len(name))
```

```python
print(len(name))
print(len(price))
print(len(og_price))
print(len(review))
print(len(rating))
print(len(discount))
print(len(des))
```

```python
df = pd.DataFrame({'Product':name,'Ratings':rating,'Review':review,'Price':price,'M.R.P':og_price,'Discount':discount,'Description':des})
df
```

```python
df.to_csv('Flipcart_WS.csv',index=False)  
```
